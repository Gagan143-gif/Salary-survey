```python
import pandas as pd
import numpy as np
import seaborn as sns #visualisation
import matplotlib.pyplot as plt #visualisation

```

```python
# Access a file from Google Drive
file_path = '/Salary_Survey.csv'
df = pd.read_csv(file_path)
```

```python
df.dtypes
```

```python
# Dropping irrelevant columns
df = df.drop(['Unnamed: 17', 'Unnamed: 18', 'Unnamed: 19', 'Unnamed: 20', 'Unnamed: 21', 'Unnamed: 22'], axis=1)
df.head()
```

```python
# Renaming the column names
df = df.rename(columns={"How old are you?":"Age","What industry do you work in?":"Work field","What city do you work in?":"City"})
df.head()
```

```python
# Total number of rows and columns
df.shape
```

```python
# Rows containing duplicate data
duplicate_rows_df = df[df.duplicated()]

```

```python
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
# Used to count the number of rows before removing the data
df.count()
```

```python
#Dropping the missing or null values
# Finding the null values.
print(df.isnull().sum())
```

```python
# Dropping the missing values.
df = df.dropna()
df.count()
```
